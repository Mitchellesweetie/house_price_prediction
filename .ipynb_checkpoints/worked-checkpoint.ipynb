```python
def is_float(x):
    try:
        float(x)
        return True
    except:
        return False
    
```

```python
df3[~df3.total_sqft.apply(is_float)].head(5)
#seeing the ranges in the first 5
```

```python
def cal_average(x):
    tokens=x.split('-')
    if len(tokens)==2:
        return(float(tokens[0])+float(tokens[0]))/2
    try:
        
        return float(x)
    except:
        return None
    
```

```python
df4=df3.copy()
df4['total_sqft1']=df4.total_sqft.apply(cal_average)
#df4.head(30)
#locating the 30th value that had ranging
df4.loc[30]
```
